```python
import pandas as pd
df=pd.read_csv("C:\\Users\\maheh\\Downloads\\Association Rules\\Association Rules\\Online retail.csv")
df
```

```python
df_expanded = df[df.columns[0]].str.split(',', expand=True)
df_expanded
```

```python
df.columns
```

```python

# Now each item will be in a separate column
trans = []
for i in range(min(7501, len(df_expanded))):  # Ensure the row index doesn't exceed the DataFrame's length
    trans.append([str(df_expanded.values[i, j]) for j in range(df_expanded.shape[1])])
trans
```

```python
pip install apyori
```

```python

from apyori import apriori
results=apriori(transactions=trans,min_support=0.003,min_confidence=0.2,min_lift=3,min_length=2,max_length=2)

report=list(results)
report
```

```python
report[0]
print(report[0][1])# Supp
print(report[0][2][0][0])
print(report[0][2][0][1])
print(report[0][2][0][2])
print(report[0][2][0][3])
```

```python
s=[]
b=[]
a=[]
c=[]
l=[]
for i in range(0,9):
    s.append(report[i][1])#supp
    b.append(report[i][2][0][0])#base
    a.append(report[i][2][0][1])#add
    c.append(report[i][2][0][2])#confi
    l.append(report[i][2][0][3])#lift

    
S_pd=pd.DataFrame(s)
B_pd=pd.DataFrame(b)
A_pd=pd.DataFrame(a)
C_pd=pd.DataFrame(c)
L_pd=pd.DataFrame(l)



```

```python

data=pd.concat([S_pd,B_pd,A_pd,C_pd,L_pd],axis=1)
data.columns = ["Base","Add",'Support','Confidence','Lift']
data
```

1).Lift: metric that measures the strength of an association between two items in a rule.
   It's important because it helps determine if the association is meaningful and worth further investigation
    ----->Lift=support/support(X)*support(Y)
2).Support: A measure of how frequently an item appears in the data.
    ----->support=Frequency(X,Y)/N
    
3).Confidence: A measure of how many times the if-then statement is found to be true.  
    ----->Confidence=Frequency(X,Y)/Frquency(x)
Limitations:
---> False Discoveries: Association rule mining can generate a large number of rules. ...
    Limited Scope: Association rule mining is primarily designed to identify binary relationships between variables,
    and may not be able to detect more complex patterns or relationships.
--->Poor performance with low support values
