```python
X=df.drop(df["Survived"])
Y=df["Survived"]
import pandas as pd
X=pd.DataFrame(X)
```

```python
#Standardisation
from sklearn.preprocessing import MinMaxScaler
MM=MinMaxScaler()
for columns in X.columns:
    if X[columns].dtype == 'float64':
        X[columns]=MM.fit_transform(X[[columns]])
X   
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
LE=LabelEncoder()
OHE=OneHotEncoder(sparse_output=False)
for columns in df.select_dtypes(include='object').columns:
    num_unique_cat=df[columns].nunique() #no.of unique cateogeories in a column
    if num_unique_cat<5:
        transformed=OHE.fit_transform(df[[columns]])
        transformed_df=pd.DataFrame(transformed,columns=OHE.get_feature_names_out([columns]))
        df=df.drop(columns,axis=1).join(transformed_df)
        
    else:
        df[columns]=LE.fit_transform(df[columns])
df       
```

```python
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR=Q3-Q1

#limits
Lower_limit=Q1-1.5*IQR
Upper_limit=Q3+1.5*IQR

outliers=((df<Lower_limit) | (df>Upper_limit))
df_clean=df[~outliers.any(axis=1)]
df_clean
```

```python
df_clean.fillna(df.mean())
```

```python
X_train=df_clean.drop(columns=["Survived"],axis=1)
Y_train=df_clean["Survived"]
```
